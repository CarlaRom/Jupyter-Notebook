### Función de pyplot que vale la pena conocer
**setparameters o setp**
Para configurar una lista de propiedades múltiples para un conjunto de líneas, puede trabajar de forma transparente con diferentes representaciones en un gráfico.

```python
lines = plt.plot(data, data, data, data**2)
```

```python
#Para cambiar propiedades globales de todo el gráfico
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, color = "r", linewidth = 2.0) #sintaxis python, con argumentos/parámetros de funciones
```

```python
#Para cambiar propiedades globales de todo el gráfico pero cambiando la sintaxis
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, "color", "r", "linewidth", 2.0) #sintaxis matlab, programación no orientada a objetos, con parámetros sin poner los = y separado por comas
```

## Múltiples gráficos en una misma figura
**numpy** trae muchas funciones para ser manipuladas y representadas gráficamente. Es una librería complementaria a la de math.

Funciones que se llaman por detrás cuando creamos varios gráficos en un dibujo (no importante aprenderlas):
Función gca (get current axis): devuelve el eje actual de dibujo, una instancia de la clase axes
Función gcf (get current figure): devuelve la figura de dibujo actual


```python
#Primero definimos la función que queremos
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5,0.1) #secuencia del 0 al 5, de 0.1 en 0.1
x2 = np.arange(0,5,0.2) #secuencia del 0 al 5, de 0.2 en 0.2
```

```python
#211: dos filas 1 columna, el gráfico va en la primera fila
#212: dos filas 1 columna, el gráfico va en la segunda fila
plt.figure(1) #para pintar en una misma figura dos gráficos (comando opcional si solo hacemos una figura)

plt.subplot(211)
plt.plot(x1,f(x1), "ro", x2, f(x2), "k") #color rojo y circulos, k para pintar la linea

plt.subplot(212)
plt.plot(x2, f(x2), "g--") #color verde y guiones

plt.show()
```

```python
#dos figuras juntas en un mismo dibujo, la primera figura con dos gráficas y la segunda con una
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

#Para volver a acceder a la primera figura que hemos hecho previamente y añadir el título:
plt.figure(1)
plt.subplot(211)
plt.title("Esto es el primer título")
```

Al renderizar las figuras se ocupa memoria que no se va a liberar a no ser que invoquemos a "close" en la ventana en la que aparece el gráfico. Es decir, tenemos que cerrar la ventana del navegador.

### CLASE TEÓRICA. FIGURAS Y EJES
¿Qué diferencia una figura de un eje?

La **figura** es la zona básica de representación, es una ventana. Una zona donde habilitamos la representación gráfica.

Dentro de la figura van los **ejes (axes)**, que conforman el rectángulo con los ejes marcados donde se va a situar el dibujo. Por cada plot que hacemos aparece una gráfica, un dibujo contenido en los ejes, que a su vez estan contenidos en la figura.

**plt.hold()**: instrucción hace que cada gráfico borre el anterior, que no se superponga sino que lo borre y que se coloque encima.
**subplot()**: permite definir zonas de dibujo, en lugar de que los dibujos se superpongan o borren el anterior, se definen subrangos de dibujo, típicamente filas y columnas. 12: 1 fila 2 columnas; 21: 2 filas 1 columna; 22: 2 filas 2 columnas.

# Textos en los gráficos
#### Añadir texto a un gráfico

```python
#Importante: Histograma de frecuencias--gráfico que funciona muy bien, informacion cuantitativa
#Generación de una distribución aleatoria
mu = 100 #media
sigma = 20 #desviación típica

x = mu + sigma * np.random.randn(10000) #generación de una campana de Gauss
```

```python
#Generación del histograma: n, número de elementos que forman parte del histograma, 
#bins, divisiones que hace de los datos para poder pintar las barras verticales (50 barritas)
#patches, cada uno de los conjuntos de datos respectivamente 
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)

plt.xlabel("Coeficiente intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$") #La r sirve para escribir strings en row (en crudo), pongo por ejemplo LaTeX
plt.text(120, 0.015, r'$\mu = 100,\ \sigma = 20$') #Añadir texto en mitad del gráfico, poniendo las coordenadas
plt.axis([20,180, 0, 0.025]) #Para asignar los límites del eje x e y                       
plt.grid(True) #Para poner una cuadrícula
plt.show()    
    
```

```python
#figsize: tamaño del gráfico en pulgadas, dpi: píxeles por pulgada
plt.figure(figsize=(10,6), dpi=90)
plt.subplot(111)

x = np.arange(0,10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw= 2.0)

plt.annotate("Máximo Local", xy = (4*np.pi,1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.ylim(-2,2)
plt.show()
```

### Cambios de ejes y escalas


```python
from matplotlib.ticker import NullFormatter #Para anular las divisiones de uno de los ejes
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0)&(y<1)] #filtramos para quedarnos solo con los y positivos y los y menores de 1
y.sort() #Números aleatorios ORDENADOS de forma CRECIENTE
x = np.arange(len(y)) #Conjunto de números de la misma longitud que y
```

```python
plt.figure(figsize = (10,8)) #Pintamos 4 gráficos en una misma figura
plt.subplot(221) #primer gráfico
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(222) #segundo gráfico
plt.plot(x,y)
plt.yscale("log") #Hay un factor 10 de diferencia entre las divisiones mayores
plt.title("Ecala Logarítmica")
plt.grid(True)

plt.subplot(223) #tercer gráfico
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthreshy=0.01) #El último término hace referencia a las divisiones de la escala logarítmica
plt.title("Escala Log Simétrico") #Hay que poner valores negativos para que nos dé un logaritmo simétrico
plt.grid(True) #Este gráfico es típico de funciones logísticas

plt.subplot(224) #cuarto gráfico
plt.plot(x,y)
plt.yscale("logit") #típico de las curvas logísticas "logit"
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) #Para cambiar el eje y, quitar el formato previo que no dejaba ver correctamente el gráfico
plt.grid(True)

#Para mover ligeramente los gráficos. hspace: espacio en altura (vertical), wspace: espacio en anchura (horizontal)
plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.25, wspace = 0.25) 

plt.show() #Hay que hacer una serie de cambios para que se vean correctamente los gráficos
```

### Ejercicio: Un ejemplo de gráfico detallado en Python
Misma función (sen, cos) en un mismo gráfico, con CAMBIOS EN LOS EJES DE DIBUJO

```python
#A continuación, generamos 256 números entre -pi y pi, separados cada uno por la misma distancia
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) #si no se coloca endpoint=True, el último dígito no se coloca, y queremos que np.pi sí se muestre
S, C = np.sin(x), np.cos(x) #Con los dígitos de antes generamos la función seno y coseno, se puede definir dos variables a la vez
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(-4,4)
plt.ylim(-1.2, 1.2)
plt.xticks(np.linspace(-4,4,9, endpoint=True)) #se pueden ajustar las marcas en el eje de las x y de las y
plt.yticks(np.linspace(-1,1,9, endpoint=True))
plt.show()
```

```python
#Modificar los límites de x e y dinámicamente para que tengan un 10% más por arriba y por abajo
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks(np.linspace(-4,4,9, endpoint=True)) #se pueden ajustar las marcas en el eje de las x y de las y
plt.yticks(np.linspace(-1,1,9, endpoint=True))
plt.show()
```

```python
plt.figure(figsize=(10,8))
#Se puede crear un vector con dónde queremos exactamente los ticks en los ejes x e y
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno")
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) #se pueden ajustar las marcas en el eje de las x y de las y, se puede poner un segundo vector con las expresiones que queremos en los ticks
plt.yticks(np.linspace(-1,1,2, endpoint=True),
          ['-1', '+1'])
#Para tener los ejes centrados en el cero, que el eje y corte por el 0, y que el eje x corte el 0 del eje y
ax = plt.gca() #Acceder a los ejes
ax.spines['right'].set_color('none') #quitar color del eje de la derecha
ax.spines['top'].set_color('none') #quitar color del eje de arriba
ax.xaxis.set_ticks_position('bottom') #poner el eje x de forma manual abajo (que ya lo está)
ax.spines['bottom'].set_position(('data', 0)) #puesto el eje x debajo, lo podemos mover al cero, poniendo 'data' como los datos originales
ax.yaxis.set_ticks_position('left') #poner el eje y a la izquierda manualmente
ax.spines['left'].set_position(('data', 0))

#Colocar una leyenda
plt.legend(loc="upper left") #loc de location


#Marcar algún punto en el gráfico
#Ejemplo el seno y coseno de 2pi/3
#Línea recta discontínua:
x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--") #1ª coordenada x0,0 y 2ª coordenada x0, sen(x0)
#Colocando el punto:
plt.scatter([x0, ], [np.sin(x0), ], 50, color="blue") #Espacio porque no necesitamos segunda coordenada
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points", #Esto coloca el texto 10 puntos a la derecha (eje x) y 30 por arriba (eje y) respecto a los datos del gráfico
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2")) #arc3: flecha ligeramente curvada

#cos:
plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--") #1ª coordenada x0,0 y 2ª coordenada x0, sen(x0)
plt.scatter([x0, ], [np.cos(x0), ], 50, color="green") 
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", #Esto coloca el texto 10 puntos a la derecha (eje x) y 30 por arriba (eje y) respecto a los datos del gráfico
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2")) 

#Para que se vean mejor las etiquetas de los ejes, es decir los números, ponemos cajitas en color blanco difuminadas por debajo
#Formato bucle
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = "r", edgecolor = "None", alpha = 0.6))

plt.show()
```

```python
plt.fill_between
```

```python
plt.figure(figsize=(8,5))
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.plot(X, Y - 1, color='blue', alpha=1.00)

plt.fill_between(X, 1, Y+1, color = "mediumpurple", alpha = 0.4)
plt.fill_between(X, -1, Y-1, color = "mediumpurple", alpha = 0.4, where = (Y-1) > -1)
plt.fill_between(X, -1, Y-1, color = "lightcoral", alpha = 0.4, where = (Y-1) < -1)

plt.xlim(-np.pi, np.pi)
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
#Pista: cuidado con la alineación del texto en el centro de las barras.
plt.figure(figsize=(8,5), dpi = 70)

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom') #ha:horizontal alignment; va:vertical alignment
for x, y in zip(X, -Y2):
    plt.text(x, y - 0.05, '%.2f' % -y, ha='center', va='top')
    
plt.ylim(-1.25, +1.25)
plt.xlim(-1, 11.5)
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
#SOLUCIÓN EJERCICIO 3 POR EL CREADOR. CÓDIGO EN GITHUB: https://github.com/rougier/matplotlib-tutorial
#import numpy as np
#import matplotlib.pyplot as plt

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5,n), plt.xticks([])
plt.ylim(-1.25,+1.25), plt.yticks([])

# savefig('../figures/bar_ex.png', dpi=48)
plt.show()
```

```python
#SOLUCIÓN EJERCICIO 4 POR EL CREADOR. CÓDIGO EN GITHUB: https://github.com/rougier/matplotlib-tutorial
#import numpy as np
#import matplotlib.pyplot as plt

n = 20
Z = np.ones(n)
Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*.05, colors=['%f' % (i/float(n)) for i in range(n)],
        wedgeprops={"linewidth": 1, "edgecolor": "black"})
plt.gca().set_aspect('equal')
plt.xticks([]), plt.yticks([])

# savefig('../figures/pie_ex.png',dpi=48)
plt.show()
```

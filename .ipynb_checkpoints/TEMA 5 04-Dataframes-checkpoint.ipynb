# CÓMO CREAR UN DATA FRAME

```python
import numpy as np
```

```python
#función de numpy random.rand() para insertar números aleatorios de 0 a 1
#Y random.randn() para insertar números aleatorios dentro de una distribución normal, con desviación típica

#Data frame con tres columnas y 10 filas
df3 = pd.DataFrame({
    'A':np.random.rand(10),
    'B':np.random.rand(10),
    'C':np.random.rand(10)
})
```

```python
df3
```

```python
df3.mean() #Calcular la media por columnas
```

```python
#Calcular la media por filas
df3.mean(axis='columns')
```

```python
df4 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'values': range(6)
},
columns = ['key', 'values'] #Para indicar el nombre de las columnas del data frame
)
df4
```

```python
df4.groupby('key') #Agrupamos los valores cualitativos de la columna 'key'
```

```python
#AGREGADO PARA OPERACIONES COMBINADAS
df4.groupby('key').mean() #Hace la media de cada grupo, es como si
                          #fuera la media de cada nivel dentro de un factor
```

```python
df4.groupby('key').median()
```

# Manipulación de data frames

```python
#Aggregate, filter, transform, apply : funciones que se pueden aplicar para operaciones combinadas
```

```python
df5 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'value1' : range(6),
    'value2' : np.random.randint(0,100,6) #6 nºs aleatorios entre 0 y 100
})
```

```python
df5
```

## Aggregate

```python
#Con aggregate podemos calcular muchos valores estadísticos de golpe
df5.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
#Utilizando un diccionario en vez de un vector
df5.groupby('key').aggregate({
    'value1' : 'min',
    'value2' : 'max'
})
```

## Filter
Eliminar datos del data frame original basándonos en una serie de propiedades

```python
df5.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std() > 13
```

```python
display('df5', "df5.groupby('key').std()", "df5.groupby('key').filter(filter_func)")
```

```python
df5.groupby('key').filter(filter_func) #Toma los datos agrupados por clave (A,B,C)
                                       #Calcula la desviación estándar de value 1 y value 2
                                       #Y filtra los datos para quedarse con los que son mayores de 27 en value 2
                                       #No devuelve el agregado
```

## Transform
Cuando queremos obtener una versión reducida de los datos pero sin combinar, solo modificamos los datos de algún modo.
Vamos a recuperar utilizando una función lambda:función definida dentro del paréntesis como parámetro de una función y que evita definir una función dentro del mismo. Conserva el tamaño del df original

```python
#Para cada clave (A,B,C) calcula la media y se la resta al dato original
#A los datos A le resta la media de los A, a los datos B le resta la media de los B y a los C le resta la media de los C.
df5.groupby('key').transform(lambda x: x-x.mean())
```

## Apply
Aplicar cualquier función arbitraria al conjunto de resultados. El parámetro siempre tiene que ser un objeto data frame y devolver un objeto pandas que sea data frame o existe el objeto pandas llamado series o bien un escalar si solo se va a devolver un número.

```python
#Normalizar la primera columna de datos según la suma de los valores de la segunda
#Versión extendida:
def norm_by_col2(x):
    x['value1'] = x['value1']/x['value2'].sum()
    return x
#Versión reducida:
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum()
    return x
```

```python
df5.groupby('key').apply(norm_by_col2)
```

# División de datos por columnas

```python
L = [0,1,0,1,2,0]
print(df5)
df5.groupby(L).sum()
#Ha juntado la que está en la posición 0,2 y 5 (por los índices de la lista con el número 0)
#También ha juntado la que está en la posición 1 y 3
#Y por último ha muestra la posición 4 que no suma con ninguna más porque no hay más de un 2
```

```python
#Ejemplo estándar agrupando por la columna 'key'
df5.groupby('key').sum()
```

```python
#Pasar todo un vector (una columna) como parámetro. Mismo resultado que antes pero formas distintas de hacerlo.
df5.groupby(df5['key']).sum()
```

```python
#Otro método: proporcionar un diccionario que indique los valores de las variables
#e índices que vamos a operar
#Elegir una columna como índice para dar nombre a las filas del df
df6 = df5.set_index('key')
df6
```

```python
#Invento de un mapeado (transformación) en forma de diccionario para cada una de las claves A,B,C
mapping = {
    'A': 'vowel',
    'B': 'consonant',
    'C': 'consonant'
}
```

```python
df6.groupby(mapping).sum()
```

```python
#Transformación de los identificadores de las filas a minúsculas
df6.groupby(str.lower).mean()
```

```python
#Agrupación por 2 factores diferentes
df6.groupby([str.lower, mapping]).mean()
```

# Gráficos desde data frames y exportación

```python
import numpy as np
years = [year for year in range (1900, 2020)]
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df7 = pd.DataFrame({
    "year": years,
    "deads": deads
})
```

3. Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio y compara el valor de los que hacen ejercicio habitualmente y los que no. ¿Observas mucha diferencia?

```python
df9.groupby(['Hace deporte']).mean()
```

```python
df9.groupby('Hace deporte').aggregate({
    'Pulso antes' : 'mean',
    'Pulso despues' : 'mean'
})
```

```python
def percentage_variation(x):
    x['Variacion'] = (((x['Pulso despues']-x['Pulso antes'])/x['Pulso antes'])*100
    return x
```

```python
df10 = df9.groupby('Hace deporte').apply(percentage_variation)
```

```python
df10.groupby('Hace deporte').mean().round(2)
#Respuesta: sí se observa diferencia, la gente que normalmente no hace deporte tiene un 
#porcentaje medio de variación del pulso mucho menor (0.21%) que la gente que hace habitualmente deporte,  
#que tiene más de un 25% de variación.
```

```python
#OTRAS RESPUESTAS:
def porce_pulso(df9):

    pmp= ((df9['Pulso despues']-df9['Pulso antes'])/df9['Pulso antes'])*100

    return pmp


print(df9.groupby('Hace deporte').apply(porce_pulso)['si'].mean())

print(df9.groupby('Hace deporte').apply(porce_pulso)['no'].mean())
```

```python
#OTRAS RESPUESTAS:
df9['var_pulso']= ((df9['Pulso despues']-df9['Pulso antes'])/df9['Pulso antes'])*100
df9.groupby('Hace deporte').aggregate('var_pulso').mean()
```
